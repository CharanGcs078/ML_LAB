```python
import pandas as pd

# Create a sample DataFrame
data = {'col1': [1, 2, 3, 4],
        'col2': ['A', 'B', 'C', 'D']}
df = pd.DataFrame(data)

# Export the DataFrame to a CSV file
df.to_csv('sample_data.csv', index=False)

print("DataFrame exported to sample_data.csv")

# Import the CSV file back into a DataFrame
df_imported = pd.read_csv('sample_data.csv')

print("\nDataFrame imported from sample_data.csv:")
df_imported
```
